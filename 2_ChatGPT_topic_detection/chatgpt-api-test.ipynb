```python
def get_topic_from_chatgpt(title):
  response = openai.ChatCompletion.create(
    model="gpt-4",#gpt-3.5-turbo",
    messages=[
          {"role": "user", "content": "what is the main topic of an article with the following title between parenthesis: (" + title +"). You must respond with one word. Avoid generic words like 'Technology'."},
      ]
  )
  return response["choices"][0]["message"]["content"]
```

```python
title = "Google Removes ‘Pirate’ URLs from Users’ Privately Saved Links"
response = get_topic_from_chatgpt(title)
response
```

```python
import pandas as pd
from tqdm import tqdm
```

```python
df = pd.read_csv("/kaggle/input/hn-titles-200-points/hn-200-points-stories-2007-02-21-to-2023-08-31.csv")
```

```python
df['topic'] = ""
df.head(3)
```

```python
for i in tqdm(range(100)):
  title = df.title.values[i]
  topic = get_topic_from_chatgpt(title)
  df.topic.values[i] = topic
```
